Problem 7

```python
#function to calculate daily water intake
def calculate_daily_water_intake(weight, activity_level):
    #Base water intake (in liters) calculation based on weight
    base_intake = weight * 0.33

    #Adjust the base intake based on activity level
    if activity_level == "low":
        return base_intake
    elif activity_level == "medium":
        return base_intake + 0.5
    elif activity_level == "high":
        return base_intake + 1.0
    else:
        return None #Invalid activity level

#Main interactive program
print("Daily Water Intake Calculator")

while True:
    try:
      #Get User weight
      weight = float(input("Enter your weight in kg: ): "))
      break
    except ValueError:
      print("Invalid input. Please enter a valid weight.")

while True:
   # Get user activity level
   activity_level = input("Enter your activity level (low/medium/high): ").lower()

   #calculate water intake
   daily_water = calculate_daily_water_intake(weight, activity_level)

   if daily_water is not None:
       print(f"Your daily water intake is approximately {daily_water:.2f} liters.")
       break
   else:
       print("Invalid activity level. Please enter 'low', 'medium', or 'high'.")

print("Stay Hydrated")

```
